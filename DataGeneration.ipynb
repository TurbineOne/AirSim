```python
objectmappingslist = open("objectmapping.txt", "r")
lines = objectmappingslist.readlines()

#We use a hash to ensure we can override mappings on multiple runs
# for instance, there is a new mapping per "play button press"

objectmappings = {}
for line in lines:
    splitsy = line.split(" ")
    objectmappings[splitsy[0]] = splitsy[1].strip()

objectmappingslist.close()

#This now holds the mapping from scene object to item class.
#The sim is able to pull which items are visible but its based on 
#scene object - which makes this mapping necessary.
print(objectmappings)
```

```python
tmp_dir = "image_capture"
print ("Saving images to %s" % tmp_dir)
try:
    os.makedirs(tmp_dir)
except OSError:
    if not os.path.isdir(tmp_dir):
        raise
```

```python
# Run through each object type and make sure they ALL have the appropriate segmentation ID
#'BP_II_C_413': 'SM_Laptop'
print("Checking object mappings...")
objecttypes = {}
#Consolidate and count keys.
for key in objectmappings:
    print(objectmappings[key])
    value = objectmappings[key]
    if value not in objecttypes:
        objecttypes[value] = 0
        
    objecttypes[value] = objecttypes[value] + 1
    
all_keys = list(objecttypes.keys())
print("Types of objects found:" + str(len(all_keys)))
```

```python
print(all_keys)
```

```python
lastThread = None
count = 0
total = 0
for key in objectmappings:
    lastThread = client.client.call_async("simSetSegmentationObjectID", key, all_keys.index(objectmappings[key]));
    count = count + 1;
    total = total + 1
    if count == 500:
        lastThread.join()
        count = 0
        print("Waiting for Unreal to catchup..." + str(total) + " of " + str(len(objectmappings.keys())))
    print(key + " set to object " + str(all_keys.index(objectmappings[key])))

#Wait only for the last thread.  This can take a while... 
lastThread.join()
```

```python
for x in range(500): # do few times
    x = random.randrange(0,2000)/100; # some random number
    y = random.randrange(0,2000)/100; # some random number
    z = 0-(random.randrange(0,700)/100); # some random number
    client.simSetVehiclePose(airsim.Pose(airsim.Vector3r(x, y, z), airsim.to_quaternion(x / 3.0, 0, x / 3.0)), True)

    responses = client.simGetImages([
        airsim.ImageRequest("front_left", airsim.ImageType.Scene),
        airsim.ImageRequest("front_right", airsim.ImageType.Scene),
        airsim.ImageRequest("front_center", airsim.ImageType.Segmentation),
        airsim.ImageRequest("front_center", airsim.ImageType.DepthPerspective),
        airsim.ImageRequest("front_center", airsim.ImageType.DepthVis),
        airsim.ImageRequest("front_center", airsim.ImageType.DisparityNormalized),
        ])

    for i, response in enumerate(responses):
        filename = os.path.join(tmp_dir, str(x) + "_" + str(y) + "_" + str(z) + "_" + str(i))
        if response.pixels_as_float:
            print("Type %d, size %d, pos %s" % (response.image_type, len(response.image_data_float), pprint.pformat(response.camera_position)))
            airsim.write_pfm(os.path.normpath(filename + '.pfm'), airsim.get_pfm_array(response))
        else:
            print("Type %d, size %d, pos %s" % (response.image_type, len(response.image_data_uint8), pprint.pformat(response.camera_position)))
            airsim.write_file(os.path.normpath(filename + '.png'), response.image_data_uint8)
```
